```python
import json
from pprint import pprint
import tinydb as tdb
```

```python
db = tdb.TinyDB('data/tinyDB.json')
```

```python
osm = db.table('osm')
merimee = db.table('merimee')
wikidata = db.table('wikidata')
```

```python
dataWiki = wikidata.all()
```

```python
dataWiki[0]
```

```python
wikiRefs = {}
for i, d in enumerate(dataWiki):
    merimeeCode = d['merimee']  # code merimee
    if merimeeCode in wikiRefs:
        wikiRefs[merimeeCode].append(i)
    else:
        wikiRefs[merimeeCode] = [i]
        
listOfDuplicates = [ (refs, indices) for refs, indices in wikiRefs.items() if len(indices)>1 ]

listOfDuplicates[:2]
```

```python
def getWikiCode(itemUrl):
    items = itemUrl.split('/')
    text = '[%s](%s)'%( items[-1], itemUrl  )
    return text

def formatWikiInfo( info ):
    code = getWikiCode( info['item'] )
    desc = info['itemDescription']
    if 'article' in info:
        pageLink = '[.w.](%s)' % info['article']
    else:
        pageLink = ''
        
    return '%s %s - %s   ' % (code, pageLink, desc)

page = ''
for merimee, indices in listOfDuplicates:
    page +=  '**%s**:   \n' % merimee
    page += '\n'.join( [formatWikiInfo(dataWiki[i]) for i in indices] )
    page += '\n\n'
    
addToClipBoard(page)
```
